```python
import json
import pandas as pd
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df_deltaA =pd.read_json("/home/crangel/PycharmProjects/EleccionesVenezuela2017/data/EDO_DELTA_AMACURO.json",lines=True)

df_deltaA['Abstencion_%'] = df_deltaA['ABSTENCION']/df_deltaA['ELECTORES INSCRITOS']
df_deltaA['winner_%'] = df_deltaA['LIZETA HERNANDEZ']/df_deltaA['VOTOS ESCRUTADOS']
df_deltaA['looser_%'] = df_deltaA['LARISSA GONZALEZ']/df_deltaA['VOTOS ESCRUTADOS']
df_deltaA
```

```python
def update_dataframe(df,PSUV,MUD):
    
    df['Abstencion_%'] = df['ABSTENCION']/df['ELECTORES INSCRITOS']
    df['PSUV_%'] = df[PSUV]/df['VOTOS ESCRUTADOS']
    df['PSUV_N'] = df[PSUV]
    df['MUD_%'] = df[MUD]/df['VOTOS ESCRUTADOS']
    df['MUD_N'] = df[MUD]
    
    unique_estados = np.unique(df['estado'])
    unique_parroquias = np.unique(df['parroquia'])
    unique_centros = np.unique(df['centro'])
    unique_municipios = np.unique(df['municipio'])
    
    average = 0
    centro_average = {}
    for i in unique_centros:
        centro = df[df['centro']==i]
        try:
            average = centro[PSUV].sum()/float(centro['VOTOS ESCRUTADOS'].sum())
        centro_average[i] = average
        
    parroquia_average = {}
    for i in unique_parroquas:
        parroquia = df[df['parroquia']==i]
        try:
            average = parroquia[PSUV].sum()/float(parroquia['VOTOS ESCRUTADOS'].sum())
        parroquia_average[i] = average
    
    municipio_average = {}
    for i in unique_municipios:
        municipio = df[df['municipio']==i]
        try:
            average = municipio[PSUV].sum()/float(municipio['VOTOS ESCRUTADOS'].sum())
        municipio_average[i] = average
        
    estado_average = {}
    for i in unique_estados:
        estado = df[df['municipio']==i]
        average = estado[PSUV].sum()/float(estado['VOTOS ESCRUTADOS'].sum())
        estado_average[i] = average
        
    print centro_average
    print parroquia_average
    print municipio_average
    print estado_average
    
    
    
    
    
```

```python
df_out = update_dataframe(df_deltaA,'LIZETA HERNANDEZ','LARISSA GONZALEZ')

df_out
```

```python
unique_parroquas = np.unique(df_deltaA['parroquia'])

parroquia1_residual = []
for i in unique_parroquas:

    parroquia1 = df_deltaA[df_deltaA['parroquia']==i]

    average = parroquia1['LIZETA HERNANDEZ'].sum()/float(parroquia1['VOTOS ESCRUTADOS'].sum())

    for i in range(0,parroquia1.shape[0]):
        
        a = parroquia1['winner_%'].iloc[i]
        parroquia1_residual.append((1-a/float(average))/0.197544083677)
    
parroquia1_residual

```
